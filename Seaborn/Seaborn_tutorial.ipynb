```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

fig, ax = plt.subplots()

# loading data
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
tips_f = np.genfromtxt(tips, delimiter=',')
# creating violinplot
ax.violinplot(tips['total_bill'], vert=False)
plt.xlabel('total_bill')

plt.show();
```

```python
# the same plot but with the use of Seaborn library
import seaborn as sns

# sns.get_dataset_names() # built-in library some datasets
tips = sns.load_dataset('tips')
# creating violinplot, label is automatically propagated to the plot,
# no need for special commend like in matplotlib
sns.violinplot(x='total_bill', data=tips)

plt.show();
```

```python
# using seaborn's colors as a colormap in matplotlib
#(moving between defaults in both libraries)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

N = 500
# constructing the colormap
current_palette = sns.color_palette('muted', n_colors=5)
print(current_palette) # list of 5(number of colors) tuples representing 3xparam color
cmap=ListedColormap(sns.color_palette(current_palette).as_hex())

# initializing the data - respectably for x axis and y axis
data1 = np.random.randn(N)
data2 = np.random.randn(N)

colors = np.random.randint(0, 5, N) # 5 possible labels for colors

plt.scatter(data1, data2, c=colors, cmap=cmap)
plt.colorbar()
plt.show();
```

```python
# controlling the plot elements with set_context() function
sns.set() # resetting default parameters

# the predefined context are: 'paper', 'notebook', 'talk', 'poster'
sns.set_context('paper', font_scale=2, rc={'font.size':2, 'axes.labelsize':12})
iris = sns.load_dataset('iris')
sns.swarmplot(x='species', y='petal_length', data=iris)
plt.show();
```

```python
# temporarily setting the plot style
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')

# setting axes style to white for first subplot; othe styles(dark, whitegrid, darkgrid, ticks)
with sns.axes_style('white'): 
    plt.subplot(211)
    sns.swarmplot(x='species', y='petal_length', data=iris)
    
# initializing second plot
plt.subplot(212)
sns.violinplot(x='total_bill', data=tips)

plt.show();
```

```python
# As for Seaborn, there are two types of functions: axes-level functions and figure-level functions. 
#The ones that operate on the Axes level are, for example, regplot(), boxplot(), kdeplot(), …,
#while the functions that operate on the Figure level are lmplot(), factorplot()->catplot(), jointplot() and a couple others.

# setting the figure size in seaborn for axes level functions
fig, ax = plt.subplots(figsize=(10,4))

iris = sns.load_dataset('iris')
sns.swarmplot(x='species', y='petal_length', data=iris, ax=ax)
plt.show();
```

```python
# for figure-level functions there have to be 2 parameters (height,aspect)
sns.set()
titanic = sns.load_dataset('titanic')
g = sns.catplot('class', 'survived', 'sex', data=titanic, kind='bar',
               height=6, aspect=2, palette='muted', legend=False)
plt.show();
```

```python
# rotating label text in seaborn
# initializing data-
x = 10**np.arange(1,10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})

# creating an lmplot
grid = sns.lmplot('x', 'y', data, height=7, truncate=True, scatter_kws={'s': 100})
grid.set_xticklabels(rotation=30) # rotating the labels on x-axis
grid.set_yticklabels(rotation=-30)
grid.set_xlabels(size=15); grid.set_ylabels(size=15)

plt.show();
```

```python
# setting xlim or ylim on seaborn
# first for boxplot, which works on Axes level
tips = sns.load_dataset('tips')
tips.head()
ax = sns.boxplot(x='total_bill', data=tips)
ax.set(xlim=(0, 100)) # or ax.set_xlom(10,100)

plt.show()
```

```python
# like above but for function at Figure-level - getting the access to Axes object
# initializing the data
x = 10 ** np.arange(1,10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})

# creating lmplot
# lmplot is a 2D scatterplot with an optional overlaid regression line.
# this is useful for comparing numeric variables
lm = sns.lmplot('x', 'y', data, size=7, truncate=True, scatter_kws={'s':100})

# getting hold of the 'Axes' objects
axes = lm.ax
axes.set_ylim(-1000000000,)

fig = lm.fig
fig.suptitle('this is the figure title', fontsize=12)

plt.show();
```

```python
# setting the log scale with data from above - with the help of Axes object

fig,ax = plt.subplots()
ax.set(xscale='log', yscale='log')

# creating a regplot - similar to lmplot but with some differences
sns.regplot('x', 'y', data, ax=ax, scatter_kws={'s':100})

plt.show();
```

```python
# for Figure level function('catplot') scaling with the set() metod of
# FacetGrid object

titanic = sns.load_dataset('titanic')
g = sns.catplot('class', 'survived', 'sex', data=titanic, kind='bar',
               size=6, palette='muted', legend=False)
# setting the 'yscale'
g.set(yscale='log')

plt.show();
```

```python
# setting title for 'Axes' level object
ax = sns.boxplot(x='total_bill', data=tips)
ax.set_title('boxplot');
```

```python
# adding title for Figure-level function
g = sns.FacetGrid(tips, col='sex', row='smoker', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip')

g.fig.suptitle('this is a title')
plt.show();
```
