```python
# working with multiple axes
fig, axes = plt.subplots(nrows=2, ncols=2) # new figure with 4 subplots
#print(fig, axes)

# giving titles for every subplots
axes[0,0].set(title='Upper Left')
axes[0,1].set(title='Upper Right')
axes[1,0].set(title='Lower Left')
axes[1,1].set(title='Lower Right')

# iterating over all items in a multidimensional numpy arrray
for ax in axes.flat: # flat attribute is for 1D array iterating purpose
    # removing all xticks and yticks
    ax.set(xticks=[], yticks=[])
plt.show();
```

```python
# the color filling between two curves
x = np.linspace(0, 10, 200)
y1 = 2 * x + 1
y2 = 3 * x + 1.2
y_mean = 0.5 * x * np.cos(2*x) + 2.5 * x + 1.1

fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, color='yellow')
ax.plot(x, y_mean, color='black') # plotting the center line

plt.show();
```

```python
# instead of the usual input data below is used a dictionary-like data object
data_obj = {'x': x,
            'y1': 2 * x + 1,
           'y2': 3 * x + 1.2,
           'mean': 0.5 * x * np.cos(2*x) + 2.5 * x + 1.1}

fig, ax = plt.subplots()
ax.fill_between('x', 'y1', 'y2', color='yellow', data=data_obj)
ax.plot('x', 'mean', color='black', data=data_obj)

plt.show();
```

```python
# exercise with matplotlib
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

# Generate data...
y_raw = np.random.randn(1000).cumsum() + 15
x_raw = np.linspace(0, 24, y_raw.size)

# Get averages of every 100 samples...
x_pos = x_raw.reshape(-1, 100).min(axis=1)
y_avg = y_raw.reshape(-1, 100).mean(axis=1)
y_err = y_raw.reshape(-1, 100).ptp(axis=1)

bar_width = x_pos[1] - x_pos[0]

# Make a made up future prediction with a fake confidence
x_pred = np.linspace(0, 30)
y_max_pred = y_avg[0] + y_err[0] + 2.3 * x_pred
y_min_pred = y_avg[0] - y_err[0] + 1.2 * x_pred

# Just so you don't have to guess at the colors...
barcolor, linecolor, fillcolor = 'wheat', 'salmon', 'lightblue'
```

```python
# Import PdfPages
from matplotlib.backends.backend_pdf import PdfPages

fig, ax = plt.subplots()
#ax.set_xlim(0, 30)
#ax.set_ylim(0, 100)

ax.plot(x_raw, y_raw, color=linecolor, linewidth=1)
ax.fill_between(x_pred, y_max_pred, y_min_pred, color=fillcolor)
ax.bar(x_pos, y_avg, width=bar_width, align='edge', color=barcolor, linewidth=1, 
       edgecolor='gray', yerr=y_err, ecolor='gray',)
ax.set(title='Future Projection of Attitudes',
       xlabel='Minutes since class began',
       ylabel='Snarkiness(snark units)')
plt.savefig('future_projection_attitudes.png')
# Initialize the pdf file
pp = PdfPages('multipage.pdf')

# Save the figure to the file
pp.savefig()
pp.close()

plt.show();
```

```python
# Input data: 2D arrays or images
# imshow, pcolor, etc - another plotting methods of matplotlib library
from matplotlib.cbook import get_sample_data
data = np.load(get_sample_data('axes_grid/bivariate_normal.npy'))

fig, ax = plt.subplots()
im = ax.imshow(data, cmap='gist_earth') #cmap - color map from mpl library
fig.colorbar(im); # plots bar with colors used in plot(ax)
```

```python
# another example with positioning of colorbar, which is a new ax in given figure
fig, ax = plt.subplots()
# positioning of the colorbar(left/righ, top/bottom, width, height)
cax = fig.add_axes([0.27, 0.8, 0.5, 0.05]) 

im = ax.imshow(data, cmap='gist_earth')
fig.colorbar(im, cax=cax, orientation='horizontal')

plt.show();
```

```python
# using vmin, vmax attributes to map with color some range of data
# data the same as above

fig, ax = plt.subplots()
im = ax.imshow(data, cmap='seismic')
fig.colorbar(im)

plt.show();
```

```python
# continuation
fig, ax = plt.subplots()
# not covered area of plot(white) in the colormap will corespond to 0
im = ax.imshow(data, cmap='seismic', vmin=-2, vmax=2)
fig.colorbar(im)
plt.show();
```

```python
# next exercise
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

plt.style.use('classic')

# Generate random data with different ranges...
data1 = np.random.random((10, 10))
data2 = 2 * np.random.random((10, 10))
data3 = 3 * np.random.random((10, 10))

# Set up our figure and axes...
fig, axes = plt.subplots(ncols=3, figsize=plt.figaspect(0.5)) # heigt/widt ratio
fig.tight_layout() # Make the subplots fill up the figure a bit more...
cax = fig.add_axes([0.25, 0.1, 0.55, 0.03]) # Add an axes for the colorbar

for ax, data in zip(axes, [data1, data2, data3]):
    im = ax.imshow(data, vmin=0, vmax=5, interpolation='nearest')
    
fig.colorbar(im, cax=cax, orientation='horizontal')

plt.show();                     
```
