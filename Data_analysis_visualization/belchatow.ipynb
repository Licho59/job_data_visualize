```python
import pandas as pd

filepath = 'Data\\belch_sept_2019.csv'
encoding = 'CP1250'
df = pd.read_csv(filepath, delimiter=';', decimal=',', encoding=encoding)
df.head(2)
```

```python
df.columns
df.Kod[0]
df.shape
df.Kod = df.Kod.str.replace('^BEL.\d-\d\d', 'BEL')
df.head(2)
```

```python
belch = df[df.Kod=='BEL']
cols = ['Doba', 'Nazwa'] + [str(i) for i in range(1,25)]
belch = belch[cols]
belch['Doba'] = pd.to_datetime(belch['Doba'].astype('str'))
belch.set_index('Doba', inplace=True)
belch.shape
belch.head()
```

```python
belch['Łącznie'] = belch.iloc[:,1:].sum(axis=1)
belch_days = belch.resample('D').sum()
belch_days.head()
```

```python
belch_month = belch.resample('M').sum()
belch_month.index.names = ['Wrzesień']
belch_month
```
