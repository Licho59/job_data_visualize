<h4>Series and DataFrame in Pandas</h4>

```python
# Series created manually
s1 = pd.Series([1,2])
s1
```

```python
s2 = pd.Series(['Boris Yeltsin', 'Mikchail Gorbachev'])
s2
```

```python
# creating DataFrame
df = pd.DataFrame([s1, s2])
df
# the same may be accomplished with a list of lists
pd.DataFrame([[1,2],['Boris Yeltsin', 'Mikchail Gorbachev']])
```

```python
# the same as above but with column labels specified
df1 = pd.DataFrame([[1,2],['Boris Yeltsin', 'Mikchail Gorbachev']],
                   columns=['column1', 'column2'])
df1
```

```python
# with row labels(the index)
df2 = pd.DataFrame([[1,2],['Boris Yeltsin', 'Mikchail Gorbachev']],
                  index=['row1', 'row2'], columns=['column1', 'column2'])
df2

print('______________________________')

# indexing DataFrame using labels
df2.loc['row1':'row2', ['column1']]
```

```python
# skipping columns argument by passing a dictionary to a DataFrame
df3 = pd.DataFrame({'column1': [1, 'Boris Yeltsin'],
                   'column2': [2, 'Mikchail Gorbachev']})
df3
```

<h4>DataFrame Math with Pandas</h4>

```python
# both ways of taking columns data 
# reviews['score']/2
m = reviews.score/2
m.head()
#reviews['score']/2

k = reviews.score/reviews['score']
k[:10]
```

```python
# boolean indexing
# counting number of scores (in DataFrame) with value equal to 10
reviews[reviews.score == 10].count()
reviews.score.max()
# counting number of scores (in Series) with value less than 3
reviews[reviews.score < 3].score.count()
# similar effect with value_counts() method
reviews[reviews.score==2].score.value_counts()
```

```python
# multiplying conditions for filtering - using parenthesis is necessary
reviews[(reviews.score > 9) & (reviews.platform == 'Xbox One')].score.value_counts()

# the same but with easier way to learn
x_box_filter = (reviews['score'] > 9) & (reviews['platform'] == 'Xbox One')
filtered_reviews = reviews[x_box_filter]
filtered_reviews.head()
```

```python
# comparing mean values for score column
reviews.score.mean()
reviews[reviews['platform']=='PlayStation 4'].score.mean()
reviews[reviews.platform == 'Xbox One'].score.mean()
```

```python
# looking for the best and the worst platform scores
mean_dict = {}
for plm in reviews.platform.unique():
    mean_dict[plm] = reviews[reviews.platform == plm].score.mean()

s = pd.Series(mean_dict)
s.head(), s.tail()

s_sorted = s.sort_values(ascending=False)
df = pd.DataFrame(s_sorted, columns=['mean_score'])

df['platform_names'] = df.index
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
index = [num for num in range(len(df))]
df = df[cols]
df.index = index
df.head()
print('5 the best game platforms:')

df.tail()
print('5 the worst game platforms')
```
